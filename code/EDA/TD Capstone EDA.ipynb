```python
import pandas as pd

def get_negative_values_table(df, df_name):
    # Filter only numeric columns
    numeric_df = df.select_dtypes(include=['int64', 'float64'])

    # Get columns with negative values and their count
    negative_counts = (numeric_df < 0).sum()

    # Filter columns with negative values
    columns_with_negative_values = negative_counts[negative_counts > 0]

    # Calculate the percentage of negative values for each column
    percentage_negative_values = (columns_with_negative_values / len(df)) * 100

    # Find the most negative (minimum) values for these columns
    most_negative_values = numeric_df[numeric_df < 0].min()

    # Filter the most negative values only for the relevant columns
    most_negative_values = most_negative_values[columns_with_negative_values.index]

    # Create a DataFrame to store the results
    negative_values_df = pd.DataFrame({
        'Negative Count': columns_with_negative_values,
        'Percentage (%)': percentage_negative_values,
        'Most Negative Value': most_negative_values
    })

    # Display the DataFrame
    print(f"Negative values in {df_name}:")
    print(negative_values_df, "\n")

# Apply to df_internal, df_external, and df_unseen (only numeric columns)
get_negative_values_table(df_internal, 'df_internal')
get_negative_values_table(df_external, 'df_external')
get_negative_values_table(df_unseen, 'df_unseen')
```

```python
import pandas as pd
import numpy as np

def display_negative_value_summary(df, df_name):
    """
    Identifies numerical columns with negative values in the DataFrame
    and displays a summary table with counts and percentages for each column.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze and display.
    df_name (str): The name of the DataFrame for display purposes.
    """

    # Step 1: Identify numerical columns with negative values
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    negative_columns = [col for col in numeric_columns if (df[col] < 0).any()]

    # Step 2: Prepare a summary table for each negative column
    summary_data = []

    for col in negative_columns:
        # Define bins with the specified range and additional sensible bins
        bins = [-99999, -99998, 0] + list(np.linspace(0.01, df[col].max(), num=10))  # Adjust num for more bins if needed

        # Use np.histogram to calculate counts
        counts, _ = np.histogram(df[col], bins=bins)

        # Calculate total and percentages
        total = sum(counts)
        percentages = [(count / total) * 100 for count in counts]

        # Prepare the summary data for the table
        summary_data.append({
            'Feature': col,
            'Counts': counts,
            'Percentages': percentages
        })

    # Step 3: Create a DataFrame for the summary
    summary_df = pd.DataFrame(summary_data)

    # Step 4: Display the summary table with the DataFrame name
    print(f"\nSummary for DataFrame: {df_name}\n")
    for index, row in summary_df.iterrows():
        print(f"\nBinned counts for {row['Feature']}:")
        summary_table = pd.DataFrame({
            'Bins': [f"{bins[i]} to {bins[i+1]}" for i in range(len(row['Counts']))],
            'Counts': row['Counts'],
            'Percentages (%)': row['Percentages']
        })
        print(summary_table)

# Example usage
display_negative_value_summary(df_internal, "df_internal")  # or df_external, "df_external", or df_unseen, "df_unseen"
display_negative_value_summary(df_external, "df_external")  # or df_external, "df_external", or df_unseen, "df_unseen"
```

For those with CC_enq = -99999,
- they also have CC_enq_L6m, CC_enq_L12m only = -99999
- and pct_CC_enq_L6m_of_L12m, pct_CC_enq_L6m_of_ever only = 0

but
- they do have CC_TL values [0 1 2]
- and a whole bunch of +ve values for CC_utilization. So they did use CC before (similar observation for those with CC_enq = 0, which is strange. If one has never enquired on CC before, why would he have utilized it?).

Answer: because credit card loans (CC_TL), utilization (CC_utilization), and flag (CC_Flag) represent the actual usage and status of credit cards, while enquiries (CC_enq) represent requests for new credit card loans. The presence of a credit card or its utilization does not necessarily mean that the person has made a recent credit card enquiry. Let’s break it down:

CC_enq (Credit Card Enquiry) = 0: This means the individual has not made any recent inquiries for a new credit card in the credit reporting period.

CC_TL (Credit Card Total Loans): Even if CC_enq is 0, this person could have existing credit card loans. For example, they may have obtained a credit card in the past, and they are still using it, but they haven't applied for a new one recently.

CC_utilization (Credit Card Utilization): Utilization refers to the percentage of the credit limit that the person is currently using. Even if they haven't made recent enquiries for new cards (CC_enq == 0), they may still be actively using their current credit card(s), hence a non-zero utilization.

```python
import numpy as np
import pandas as pd

# Function to display binned frequency table for percentage-based columns
def display_binned_frequency_summary(df, columns, df_name):
    """
    Displays a binned frequency summary table for the specified columns in the DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.
    columns (list): List of columns to analyze.
    df_name (str): The name of the DataFrame for display purposes.
    """

    # Step 1: Prepare a summary table for each specified column
    summary_data = []

    for col in columns:
        # Define bins from 0 to the maximum value of the column
        bins = [-99999, -99998, 0, 0.01, 100] + list(np.linspace(100.01, df[col].max(), num=10))  # Adjust num for more bins if needed

        # Use np.histogram to calculate counts
        counts, _ = np.histogram(df[col], bins=bins)

        # Calculate total and percentages
        total = sum(counts)
        percentages = [(count / total) * 100 if total > 0 else 0 for count in counts]

        # Prepare the summary data for the table
        summary_data.append({
            'Feature': col,
            'Counts': counts,
            'Percentages': percentages,
            'Bins': bins
        })

    # Step 2: Display the summary table with the DataFrame name
    print(f"\nBinned Frequency Summary for DataFrame: {df_name}\n")

    for index, row in enumerate(summary_data):
        print(f"\nBinned counts for {row['Feature']}:")

        # Create a DataFrame to display the bins, counts, and percentages
        summary_table = pd.DataFrame({
            'Bins': [f"{row['Bins'][i]} to {row['Bins'][i+1]}" for i in range(len(row['Counts']))],
            'Counts': row['Counts'],
            'Percentages (%)': row['Percentages']
        })

        print(summary_table)
        print("\n" + "-"*50)  # Separator for clarity

# Step 3: Apply the function to the filtered percentage columns
display_binned_frequency_summary(df_merged, filtered_pct_columns, "df_merged")
```

if a loan flag is 0, would Approved_Flag have no values? Appears not. Will have P1-4 regardless.

Because CC_Flag etc just means whether the person has an active credit card loan etc. One may have no active loans, but has a high priority (= low risk) to be approved. See below...

* CC_Flag:
Meaning: This flag indicates **whether the individual has an active credit card account.**
Possible Values: It may be a binary value (e.g., 0 = No active credit card, 1 = Has an active credit card).

* PL_Flag:
Meaning: This flag denotes **whether the individual has a personal loan.**
Possible Values: Again, it is typically binary (e.g., 0 = No personal loan, 1 = Has a personal loan).

* HL_Flag:
Meaning: This flag represents **whether the individual has a home loan or mortgage.**
Possible Values: Usually binary (e.g., 0 = No home loan, 1 = Has a home loan).

* GL_Flag:
Meaning: This flag indicates **whether the individual has a gold loan, which is a loan secured by gold assets.**
Possible Values: Often binary (e.g., 0 = No gold loan, 1 = Has a gold loan).

* Approved_Flag. :
  * P1:
Meaning: **Approved as a low-risk loan**. This might indicate that the applicant has a strong credit profile and meets all the criteria for an ideal candidate.
  * P2:
Meaning: **Approved as a medium-risk loan**. This could suggest that the applicant is acceptable but may have some issues or less favorable credit attributes compared to those in the P1 category.
  * P3:
Meaning: **Approved as a high-risk loan**. This may indicate that while the loan is approved, the applicant has a higher risk profile and may face higher interest rates or stricter terms.
  * P4:
Meaning: **Application is pending or requires additional review**. This could imply that the application is under consideration and may need further assessment or documentation before a decision is made.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create income bins and labels
income_bins = [0, 15000, 30000, 45000, 60000, 100000, float('inf')]
income_labels = ['0-15k', '15k-30k', '30k-45k', '45k-60k', '60k-100k', '100k+']
df_merged['NETMONTHLYINCOME_BINNED'] = pd.cut(df_merged['NETMONTHLYINCOME'], bins=income_bins, labels=income_labels)

# Create age bins and labels
age_bins = [0, 18, 30, 45, 60, 75, float('inf')]
age_labels = ['0-18', '19-30', '31-45', '46-60', '61-75', '76+']
df_merged['AGE_BINNED'] = pd.cut(df_merged['AGE'], bins=age_bins, labels=age_labels)

# List of new categorical columns, adding binned age column
categorical_columns = ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'NETMONTHLYINCOME_BINNED', 'AGE_BINNED']

# Create a figure and subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))  # 3 rows, 2 columns (2 extra spaces)
axes = axes.flatten()  # Flatten to easily iterate over

# Loop through each categorical variable and its corresponding subplot
for i, column in enumerate(categorical_columns):
    # Calculate counts for each category
    counts = df_merged[column].value_counts().sort_index()

    # Create bar plot
    bars = axes[i].bar(counts.index.astype(str), counts.values, color='skyblue', edgecolor='black', width=0.4)  # Set width smaller to prevent overlap

    # Calculate total and percentages for each bar
    total = counts.sum()  # Total count of all records
    percentages = [(count / total) * 100 for count in counts.values]  # List of percentages for each category

    # Annotate bars with percentages
    for bar, percentage in zip(bars, percentages):
        height = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width() / 2, height, f'{percentage:.2f}%',
                      ha='center', va='bottom')  # Center the percentage above the bar

    # Set x-ticks to spread evenly
    axes[i].set_xticks(range(len(counts)))  # Use range to evenly space the ticks
    axes[i].set_xticklabels(counts.index.astype(str))  # Set tick labels to string

    # Rotate x-tick labels
    plt.setp(axes[i].xaxis.get_majorticklabels(), rotation=45, ha="right")

    # Set titles and labels
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel('Category')
    axes[i].set_ylabel('Frequency')

# Hide the unused subplot (6th one)
axes[-1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Correlation between variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


# Assuming df_merged is your merged DataFrame
# Step 1: Extract all continuous features
continuous_features = df_merged.select_dtypes(include=['int64', 'float64'])

# Step 2: Calculate the correlation matrix
correlation_matrix = continuous_features.corr()

# Step 3: Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Step 3: Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create the heatmap
sns.heatmap(correlation_matrix, annot=False, fmt='.2f', cmap='coolwarm', mask=mask,
            square=True, linewidths=0.5, cbar_kws={"shrink": .8})

# Set titles and labels
plt.title('Correlation Heatmap of Continuous Variables', fontsize=16)

# # Adjust the x-tick and y-tick labels
# plt.xticks(rotation=90, fontsize=10)  # Set fontsize for x-tick labels
# plt.yticks(rotation=0, fontsize=10)    # Set fontsize for y-tick labels

# Get the tick positions and labels
tick_labels = correlation_matrix.columns
# Set all x-tick and y-tick labels
plt.xticks(ticks=range(len(tick_labels)), labels=tick_labels, rotation=90, fontsize=6, ha='left')  # Set fontsize for x-tick labels
plt.yticks(ticks=range(len(tick_labels)), labels=tick_labels, rotation=0, fontsize=6, va='top')    # Set fontsize for y-tick labels

# Show the plot
plt.tight_layout()
plt.show()
```
